```python
from statistics import correlation

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df = pd.read_csv("../data/silver/hmeq_silver.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="BAD", data=df)
plt.title("Distribuição da variável alvo (BAD)")
plt.show()
```

```python
percent = df["BAD"].value_counts(normalize=True) * 100
print(percent)
```

```python
correlation = df.corr()
sns.heatmap(correlation, cmap="coolwarm", annot=False)
plt.title("Matriz de Correlação")
plt.show()
```

```python
sns.boxplot(x="BAD", y="LOAN", data=df)
plt.title("Empréstimos por situação de pagamento")
plt.show()
```

```python
sns.boxplot(x="BAD", y="DEBTINC", data=df)
plt.title("Índice de Endividamento  por situação de pagamento")
plt.show()
```
